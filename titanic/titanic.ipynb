```python
females = pd.DataFrame(data[data['Sex']=='female']['Name'])
females
```

```python
females[['Last Name', 'Name']] = females['Name'].str.split(',', expand = True)
females
```

```python
females[['Title', 'Name']] = females['Name'].str.strip().str.split(' ', expand = True, n=1)
females
```

```python
females['Title'].value_counts()
```

```python
female_names = {}
```

```python
females[females['Title']=='Miss.'].head(60)
```

```python
for s in females[females['Title']=='Miss.']['Name']:
    s = s.strip('()')
    '''
    t = s.find('(')
    if t == -1:
        t = s.find('"')
    names = s.split()
    if t != -1:
        names = s[:t].split()
    '''
    names = [s.split()[0]]
    for n in names:
        if n in female_names:
            female_names[n] += 1
        else:
            female_names[n] = 1        
female_names
```

```python
females[females['Title']=='Mrs.'].head(60)
```

```python
for s in females[females['Title']=='Mrs.']['Name']:
    s = s.strip('()')
    t = s.find('(')
    names = s.split()
    if t != -1:
        names = s[t+1:].split()
    names = [names[0]]
    #if len(names) > 1:
        #names = names[:-1]
    for n in names:
        if n in female_names:
            female_names[n] += 1
        else:
            female_names[n] = 1        
female_names
```
