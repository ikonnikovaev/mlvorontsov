```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
prices = pd.read_csv('close_prices.csv', index_col='date')
prices
```

```python
djia = pd.read_csv('djia_index.csv', index_col='date')
djia
```

```python
pca = PCA(n_components=10)
```

```python
pca.fit(prices)
```

```python
evr = pca.explained_variance_ratio_ 
evr
```

```python
s = 0
k = 0
while k < evr.size and s < 0.9:
    s += evr[k]
    k += 1
k
```

```python
out = open('q1.txt', 'w')
out.write(str(k))
out.close()
```

```python
components = pca.components_
components.shape
```

```python
company = np.argmax(components[0])
company = prices.columns[company]
company
```

```python
transformed_prices = pca.transform(prices)
transformed_prices.shape
```

```python
djia['prices'] = transformed_prices[:, 0]
djia
```

```python
c = djia['^DJI'].corr(djia['prices'])
c
```

```python
out = open('q2.txt', 'w')
out.write(str(c))
out.close()
```
