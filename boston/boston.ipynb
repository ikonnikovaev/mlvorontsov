```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import preprocessing
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsRegressor
```

```python
df = datasets.load_boston()
X = df.data
y = df.target
```

```python
X = preprocessing.scale(X)
```

```python
kf = KFold(n_splits=5,shuffle=True, random_state=42)
```

```python
p_opt = 0
max_score = 0
for p_cur in np.linspace(1, 10, num=200):
    reg = KNeighborsRegressor(n_neighbors=5, weights='distance', metric='minkowski', p=p_cur)
    score = cross_val_score(reg, X, y, cv=kf, scoring='neg_mean_squared_error').mean()
    print(p_cur)
    print(score)
    if p_cur == 1 or score > max_score:
        max_score = score
        p_opt = p_cur
```
