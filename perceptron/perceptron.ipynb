```python
import pandas as pd
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
data_train = pd.read_csv('perceptron-train.csv', header=None)
data_train
```

```python
data_test = pd.read_csv('perceptron-test.csv', header=None)
data_test
```

```python
X_train = data_train.drop(columns= [0])
y_train = data_train[0]
X_train
```

```python
X_test = data_test.drop(columns=[0])
y_test = data_test[0]
X_test
```

```python
clf = Perceptron()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
sc0 = accuracy_score(y_test, y_pred)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
clf.fit(X_train_scaled, y_train)
y_pred = clf.predict(X_test_scaled)
sc = accuracy_score(y_test, y_pred)
```

```python
f = open('q1.txt', 'w')
f.write(str(sc - sc0))
f.close()
```
